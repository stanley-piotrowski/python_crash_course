## Plotting a simple line graph


```python
# Plot square numbers
squares = []
for num in range(1, 11):
    squares.append(num ** 2)

# Create plot
%matplotlib inline # will be used for all plots after
matplotlib.pyplot.plot(squares)
matplotlib.pyplot.show()
```

```python
# Customize visualization
# Add input values-- matplotlib assumes a sequence of numbers should start at 0, not 1
input_values = list(range(1, len(squares) + 1))
matplotlib.pyplot.plot(input_values, squares, linewidth = 5)

# Set title and label axes
matplotlib.pyplot.title('Square numbers', fontsize=24)
matplotlib.pyplot.xlabel('Value', fontsize=14)
matplotlib.pyplot.ylabel('Square of value', fontsize=14)

# Set size of axis ticks
matplotlib.pyplot.tick_params(axis='both', labelsize=14)

# View inline plot
matplotlib.pyplot.show()
```
